#### After visualizing the data in Retool, I realized there were some unit errors in the dataset. Therefore, we are modifying the units according to the type of project (generation, storage, or grid).

```python
##

sample_projects_df

unique_values = sample_projects_df['capacity_units'].unique()

type_to_unit = {
    'generation': 'MW',
    'storage': 'MWh',
    'grid': 'KV'
}

# Ensure type is lowercase to match mapping
sample_projects_df['type_lower'] = sample_projects_df['type'].str.lower()

# Create a new Series of updated units from the mapping
new_units = sample_projects_df['type_lower'].map(type_to_unit)

# Only update where a match was found in the mapping
sample_projects_df['capacity_units'] = new_units.combine_first(sample_projects_df['capacity_units'])

# Optional: drop the helper column
sample_projects_df.drop(columns='type_lower', inplace=True)

sample_projects_df
```
